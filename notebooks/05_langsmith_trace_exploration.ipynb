# LangSmith Trace Exploration

LangSmith is the primary observability surface for this project.

The goal of this notebook is to connect what you see in Python to what you should expect to see in the LangSmith UI.

## Step 1: Run A Small Traced Experiment

This cell runs a few campaigns from the saved dataset through the agent.

Why this matters:

- each campaign run should appear as one parent trace in LangSmith
- the saved dataset means you can match a trace back to a specific campaign row on disk
- keeping the sample small makes the UI easier to inspect while learning


```python
from iroas_agent.runner import run_experiment
```

```python
output = run_experiment(dataset_path='data/synthetic_campaigns.csv', sample_size=3)
[(row['campaign_id'], row['metadata']) for row in output['results'][:3]]
```

## Step 2: Inspect The Metadata We Attach To Each Run

This cell pulls one example metadata block from the Python results.

Why this matters:

- these are the fields you should expect to see attached to runs in LangSmith
- they make filtering and debugging much easier
- they are especially useful when you want to compare behavior across data availability slices or estimator choices


```python
output['results'][0]['metadata']
```

## Step 3: Inspect A Full Trajectory In Python First

Before opening LangSmith, it helps to look at the trajectory directly in the notebook.

Why this matters:

- it gives you a concrete expectation for what the trace tree should contain
- it makes it easier to distinguish parent run metadata from child tool spans
- it reinforces that the trace is not just logging, but a structured view of the agent policy


```python
for step in output['results'][0]['trajectory']:
    print('Thought:', step['thought'])
    print('Action:', step['action'])
    print('Observation:', step['observation'])
    print('-' * 80)
```

## Step 4: What To Look For In LangSmith

When you open the matching run in LangSmith, verify that you can see:

- one trace per campaign
- child spans for tool calls
- metadata such as `campaign_id`, `has_lift_test`, `has_geo_experiment`, `predicted_iROAS`, and `true_iROAS`
- tags for final estimator used, number of steps, and data availability type

That is the observability contract this project is trying to teach.
